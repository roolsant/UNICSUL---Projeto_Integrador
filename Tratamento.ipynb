##### Tratamento para a tabela tb_Circuitos

```python
import pandas as pd
import os

pais_csv = 'C:/Users/rodri/OneDrive - 5pwtcn/Cruzeiro_do_Sul_Virtual/Projeto_Integrador/Atividade_2/Extracoes_SQL/tb_pais.csv'
circuits_csv ='C:/Users/rodri/OneDrive - 5pwtcn/Cruzeiro_do_Sul_Virtual/Projeto_Integrador/Atividade_2/Fontes/Kaggle/F1_Championship/circuits.csv'
circuits_length_csv = 'C:/Users/rodri/OneDrive - 5pwtcn/Cruzeiro_do_Sul_Virtual/Projeto_Integrador/Atividade_2/Fontes/Circuits_Length.csv'

try:
       
    ### carregando o arquivo tb_Pais jÃ¡ com os Ids do banco
    df_pais = pd.read_csv(pais_csv)

    ### carregando o arquivo circuits com os dados dos circuitos
    df_circuitos_original = pd.read_csv(circuits_csv)

    ### carregando o arquivo circuits com as extensÃµes das pistas
    df_extensoes = pd.read_csv(circuits_length_csv, delimiter=';')

  
    ### juntando os dados de circuitos com as extensÃµes
    ### usando o 'left_on' e 'right_on' porque os nomes das colunas de junÃ§Ã£o sÃ£o diferentes
    df_merged = pd.merge(
        left=df_circuitos_original,
        right=df_extensoes,
        left_on='name',
        right_on='Circuit',
        how='left'
    )

    ### fazendo a padronizacao dos nomes dos paÃ­ses
    correcoes_paises = {
        'UK': 'United Kingdom',
        'USA': 'United States',
        'UAE': 'United Arab Emirates',
        'Korea': 'South Korea',
        'Russia': 'Russian Federation'
    }
    df_merged['country_padronizado'] = df_merged['country'].replace(correcoes_paises)

    ### criando o dicionÃ¡rio de mapeamento: {'nm_pais': id_pais}
    mapa_pais_para_id = pd.Series(df_pais['idpais'].values, index=df_pais['nm_pais']).to_dict()
    
    ### usando o mapa para criar a nova coluna de chave estrangeira
    df_merged['tb_Pais_idpais'] = df_merged['country_padronizado'].map(mapa_pais_para_id)

    ### selecionando apenas as colunas para a tabela final.
    df_final_circuito = df_merged[['name', 'Length KM', 'tb_Pais_idpais']].copy()
    
    ### renomeando as colunas para bater com a tabela do banco de dados.
    df_final_circuito.rename(columns={
        'name': 'nm_circuito',
        'Length KM': 'nr_extensao'
    }, inplace=True)
    
    ### salvando o arquivo CSV final
    pasta_saida = 'C:/Users/rodri/OneDrive - 5pwtcn/Cruzeiro_do_Sul_Virtual/Projeto_Integrador/Atividade_2/Tabelas'
    nome_arquivo = 'tb_Circuito.csv'
    caminho_completo = os.path.join(pasta_saida, nome_arquivo)
    
    os.makedirs(pasta_saida, exist_ok=True)
    
    df_final_circuito.to_csv(caminho_completo, index=False, encoding='utf-8-sig')
    
    print(f"\nâœ… Sucesso! O arquivo '{caminho_completo}' foi salvo e estÃ¡ pronto para a carga no banco.")


except FileNotFoundError as e:
    print(f"ðŸš¨ ERRO: Arquivo nÃ£o foi encontrado. Verifique os caminhos e nomes dos arquivos. Detalhe: {e}")
except KeyError as e:
    print(f"ðŸš¨ ERRO de Chave: Uma coluna esperada nÃ£o foi encontrada. Verifique os nomes das colunas nos arquivos CSV. Detalhe: {e}")
except Exception as e:
    print(f"ðŸš¨ Ocorreu um erro inesperado: {e}")
```

##### Tratamento para a tabela tb_Provas

```python
import pandas as pd
import os

try:
    # --- Passo 1: Carregar as fontes de dados ---
    circuitos_db_csv = 'C:/Users/rodri/OneDrive - 5pwtcn/Cruzeiro_do_Sul_Virtual/Projeto_Integrador/Atividade_2/Extracoes_SQL/tb_ciruito.csv'
    races_csv ='C:/Users/rodri/OneDrive - 5pwtcn/Cruzeiro_do_Sul_Virtual/Projeto_Integrador/Atividade_2/Fontes/Kaggle/F1_Championship/races.csv'
    circuitosKaggle_csv = 'C:/Users/rodri/OneDrive - 5pwtcn/Cruzeiro_do_Sul_Virtual/Projeto_Integrador/Atividade_2/Fontes/Kaggle/F1_Championship/circuits.csv'
    
    df_races = pd.read_csv(races_csv)
    df_circuitos_db = pd.read_csv(circuitos_db_csv) 
    df_circuitos_kaggle = pd.read_csv(circuitosKaggle_csv)
    
    print("--- Arquivos carregados ---")

    # --- Passo 2: Mapa de CorreÃ§Ã£o de Nomes (incluindo limpeza de strings) ---
    df_circuitos_kaggle['join_key'] = df_circuitos_kaggle['name'].str.lower().str.strip()
    df_circuitos_db['join_key'] = df_circuitos_db['nm_circuito'].str.lower().str.strip()
    
    # --- Passo 3: Criar o Mapeamento de IDs ---
    ponte_id_nome = df_circuitos_kaggle[['circuitId', 'join_key']]
    mapa_df = pd.merge(
        ponte_id_nome,
        df_circuitos_db[['idcircuito', 'join_key']],
        on='join_key'
    )
    mapa_final_ids = mapa_df[['circuitId', 'idcircuito']]
    
    # --- Passo 4: Juntar o mapa com os dados das corridas ---
    df_provas_com_id_final = pd.merge(
        df_races,
        mapa_final_ids,
        on='circuitId',
        how='left'
    )
    
    # --- Passo 5 (NOVO): Remover as linhas que nÃ£o foram mapeadas ---
    linhas_antes = len(df_provas_com_id_final)
    # .dropna() remove as linhas onde a coluna 'idcircuito' Ã© Nula/Vazia
    df_provas_com_id_final.dropna(subset=['idcircuito'], inplace=True)
    linhas_depois = len(df_provas_com_id_final)
    
    print(f"\nLimpeza de registros nÃ£o mapeados concluÃ­da.")
    print(f"Total de linhas original: {linhas_antes}")
    print(f"Total de linhas removidas: {linhas_antes - linhas_depois}")
    print(f"Total de linhas final: {linhas_depois}")

    # --- Passo 6: Formatar as colunas restantes ---
    df_provas_com_id_final['time'] = df_provas_com_id_final['time'].replace('\\N', '00:00:00')
    df_provas_com_id_final['dt_prova'] = pd.to_datetime(df_provas_com_id_final['date'] + ' ' + df_provas_com_id_final['time'], errors='coerce')
    df_provas_com_id_final['nm_situacao'] = 'Finished'

    # --- Passo 7: Selecionar, Renomear e Salvar o arquivo final ---
    df_final_prova = df_provas_com_id_final[['dt_prova', 'nm_situacao', 'idcircuito']].copy()
    df_final_prova.rename(columns={'idcircuito': 'id_circuito'}, inplace=True)
    df_final_prova['id_circuito'] = df_final_prova['id_circuito'].astype('Int64')
    
    pasta_saida = 'C:/Users/rodri/OneDrive - 5pwtcn/Cruzeiro_do_Sul_Virtual/Projeto_Integrador/Atividade_2/Tabelas'
    nome_arquivo = 'tb_Prova.csv'
    caminho_completo = os.path.join(pasta_saida, nome_arquivo)
    os.makedirs(pasta_saida, exist_ok=True)
    df_final_prova.to_csv(caminho_completo, index=False, encoding='utf-8-sig')
    
    print(f"\nâœ… Sucesso! O arquivo '{caminho_completo}' foi salvo e estÃ¡ 100% limpo.")

except Exception as e:
    print(f"ðŸš¨ Ocorreu um erro inesperado: {e}")
```
